```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
df = pd.read_csv(url)
df.head()
```

```python
india_df = df[df['Country Name'] == 'India']
india_df = india_df[india_df['Year'] >= 2000]
india_df.head()
```

```python
plt.figure()
plt.bar(india_df['Year'], india_df['Value'])

plt.title("India Population Growth (2000 onwards)")
plt.xlabel("Year")
plt.ylabel("Population")

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.hist(india_df['Value'], bins=10)

plt.title("Population Distribution of India")
plt.xlabel("Population Range")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure()
sns.lineplot(x=india_df['Year'], y=india_df['Value'])

plt.title("India Population Trend (Stylish View)")
plt.xlabel("Year")
plt.ylabel("Population")

plt.show()
```
